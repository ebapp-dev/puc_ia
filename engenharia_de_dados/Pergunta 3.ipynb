##Média de crianças por organização / tipo de evento / mês

- Crianças (dim_children)
- Organização (dim_children, dim_events)
- Mês (dim_events)
- Tipo de evento (dim_events)

```
#Apaga todas as tabelas do level3 diretório
dbutils.fs.rm('dbfs:/user/hive/warehouse/level3.db/tb_fato3', True)
```

```
%sql

USE level3;
DROP TABLE IF EXISTS tb_fato3;
CREATE TABLE tb_fato3 AS
SELECT org, mes, ano, tipo, CAST(avg(quant) over
(order by org, dia, mes, ano, tipo rows unbounded preceding) AS decimal(5,2)) as avg
FROM (SELECT A.org, extract(DAY FROM B.evento) as dia, extract(MONTH FROM B.evento) as mes, extract(YEAR FROM B.evento) as ano, B.eventoTipo as tipo, count(*) as quant 
FROM dim_events B
INNER JOIN dim_children A ON (A.org = B.org AND A.mat = B.mat)
GROUP BY A.org, dia, mes, ano, eventoTipo
ORDER BY org, tipo, ano, mes, dia);

SELECT * FROM tb_fato3;
```
