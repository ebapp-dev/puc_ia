##Média de crianças por organização / classe / mês

- Crianças (dim_children)
- Organização (dim_children, dim_events)
- Mês (dim_events)
- Classe (dim_classes)

```
#Apaga todas as tabelas do level3 diretório
dbutils.fs.rm('dbfs:/user/hive/warehouse/level3.db/tb_fato4', True)
```

```
%sql

USE level3;
DROP TABLE IF EXISTS tb_fato4;
CREATE TABLE tb_fato4 AS
SELECT org, mes, ano, classe, cast(avg(quant) over
(order by org, dia, mes, ano, classe rows unbounded preceding) AS decimal(5,2)) as media
FROM (SELECT A.org, extract(DAY FROM B.evento) as dia, extract(MONTH FROM B.evento) as mes, extract(YEAR FROM B.evento) as ano, C.classe as classe, count(*) as quant 
FROM dim_events B
INNER JOIN dim_children A ON (A.org = B.org AND A.mat = B.mat)
INNER JOIN dim_classes C ON (A.idade >= C.idademin and A.idade <= C.idademax)
GROUP BY A.org, dia, mes, ano, classe
ORDER BY org, ano, mes, dia, classe);

SELECT * FROM tb_fato4;
```
