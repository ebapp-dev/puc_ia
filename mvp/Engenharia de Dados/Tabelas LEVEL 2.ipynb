###Tabela tb_events
A tabela tb_events precisa de uma coluna que identifique o tipo de evento. Os eventos são classificados pelo dia e hora que são realizados. Essa classificação é constante, ou seja, todas as organzações relizam 2 eventos aos domingos e um na quarta.
1 = domingo pela amanhã, 2 = domingo à noite, 3 = quarta à noite.

Obs: Todos os registros que não obedecem essa regra e são da organização teste serão eliminados.

```
%sql

DROP TABLE IF EXISTS level2.`tb_events_l2`;

-- Cria nova coluna eventoTipo
CREATE TABLE level2.`tb_events_l2` AS
SELECT *, IF(EXTRACT(DAYOFWEEK FROM evento) > 1, 3,IF(EXTRACT(HOUR FROM evento) < 12, 1, 2)) as eventoTipo
FROM level1.`tb_events`;

-- Elimina registros com org = teste e eventos que não foram realizados domingo ou quarta
DELETE FROM level2.`tb_events_l2`
WHERE EXTRACT(DAYOFWEEK FROM evento) NOT IN (1,4) OR org = 'teste';

select * from level2.tb_events_l2

```

###Tabela tb_parents
Serão eliminados os registros org = teste e elimina espaços causados pela digitação no campo 'bairro'.

```
%sql

-- A tabela parents precisa de limpeza nos dados do bairro que apresenta espaços no final devido a digitação

-- Cria database para receber as tabelas transformadas

DROP TABLE IF EXISTS level2.`tb_parents_l2`;

CREATE TABLE level2.`tb_parents_l2` AS
SELECT org, matResp, trim(bairro) as bairro, cep, cidade 
FROM level1.`tb_parents`;

-- Elimina registros com org = teste
DELETE FROM level2.`tb_parents_l2`
WHERE org = 'teste';

select * from level2.tb_parents_l2;
```

###Tabela tb_children
Inserir nova coluna com a idade da criança e eliminar os registros org = teste.

```
%sql

-- Cria a coluna 'idade'
use level2;
DROP TABLE IF EXISTS level2.`tb_children_l2`;
CREATE TABLE tb_children_l2 AS
SELECT *, datediff(YEAR, dtNasc, now() ) as idade
FROM level1.`tb_children`;

select * from tb_children_l2;
```

```
%sql
-- Limpeza de registros de tb_children_l2
USE level2;
DELETE FROM tb_children_l2
WHERE org = 'teste';
```
